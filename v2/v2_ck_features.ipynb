```python
# import required packages
import pandas as pd
```

```python
# open csv with meditative states
ck_med = pd.read_excel('CK_med.xlsx', index_col=None, header=0)
ck_med['label'] = 'ck_med'
ck_med = ck_med.drop(labels = ['Total Power','SDNN'], axis = 1)
ck_med
```

```python
# open csv with non-meditative states
ck_pre = pd.read_excel('CK_pre.xlsx', index_col=None, header=0)
ck_pre['label'] = 'ck_pre'
ck_pre = ck_pre.drop(labels = ['Total Power','SDNN'], axis = 1)
ck_pre
```

```python
df = pd.concat([ck_pre,ck_med])
df
```

```python
# save df into a csv file
df.to_csv('v2_ck_features.csv',index=False)
```

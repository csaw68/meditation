```python
# import required packages
import pandas as pd
```

```python
# open csv with Chi meditative states
med = pd.read_excel('K_med.xlsx', index_col=None, header=0)
med['label'] = 'k_med'
med = med.drop(labels = ['Total Power','SDNN'], axis = 1)
med
```

```python
# open csv with non-meditative states
pre = pd.read_excel('K_pre.xlsx', index_col=None, header=0)
pre['label'] = 'k_pre'
pre = pre.drop(labels = ['Total Power','SDNN'], axis = 1)
pre
```

```python
df = pd.concat([pre,med])
df
```

```python
# save df into a csv file
df.to_csv('v2_k_features.csv',index=False)
```
